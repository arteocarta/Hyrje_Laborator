# Eksperimenti A – Matja e fushës magnetike ambientale

### Qëllimi
Të matet fusha magnetike në një vend pa magnet të afërt dhe të vlerësohet stabiliteti i leximit të sensorit.


```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt

df_bg = pd.read_csv("data/B_background.csv")

Bx_bg = df_bg["Bx_uT"].to_numpy()
By_bg = df_bg["By_uT"].to_numpy()
Bz_bg = df_bg["Bz_uT"].to_numpy()

B_bg = np.sqrt(Bx_bg**2 + By_bg**2 + Bz_bg**2)

print("B mesatare =", np.mean(B_bg))
print("Devijimi standard =", np.std(B_bg, ddof=1))

plt.plot(B_bg)
plt.xlabel("Indeksi i matjes"); plt.ylabel("B [uT]")
plt.title("Fusha magnetike ambientale në kohë")
plt.grid(True); plt.show()
```

### Interpretimi
- Fusha magnetike ambientale matet mesatarisht rreth 44.7 µT, brenda vlerave tipike për zonën tonë gjeografike.  
- Devijimi standard i vogël tregon stabilitet të lartë të sensorit.  
- Kjo vlerë bazë do të përdoret si referencë sfond (background) në analizat vijuese.


# Eksperimenti B – Hartëzimi 1D pranë magnetit

### Qëllimi
Të shihet si ndryshon fusha magnetike me largësinë nga magneti.


```python
df1 = pd.read_csv("data/B_1D.csv")

df1["B_uT"] = np.sqrt(df1["Bx_uT"]**2 + df1["By_uT"]**2 + df1["Bz_uT"]**2)
df1["DeltaB_uT"] = df1["B_uT"] - np.mean(B_bg)

plt.plot(df1["x_cm"], df1["DeltaB_uT"], "o-")
plt.xlabel("Largësia x [cm]"); plt.ylabel("Delta B [uT]")
plt.title("Perturbimi magnetik si funksion i largësisë")
plt.grid(True); plt.show()
```

### Interpretimi
- Intensiteti i fushës bie qartë me rritjen e largësisë nga magneti.  
- Profili i rënies përputhet me ligjin e dipolit magnetik ($\Delta B \propto 1/r^3$).  
- Në largësi mbi 15 cm, vlera e fushës konvergon me sfondin ambiental.


# Eksperimenti C – Hartëzimi 2D i fushës magnetike

### Qëllimi
Të ndërtohet një hartë 2D e intensitetit të fushës magnetike pranë magnetit.


```python
df2 = pd.read_csv("data/B_map_2D.csv")

df2["B_uT"] = np.sqrt(df2["Bx_uT"]**2 + df2["By_uT"]**2 + df2["Bz_uT"]**2)
df2["DeltaB_uT"] = df2["B_uT"] - np.mean(B_bg)

plt.figure(figsize=(6,5))
plt.scatter(df2["x_cm"], df2["y_cm"], c=df2["DeltaB_uT"], s=120)
plt.colorbar(label="Delta B [uT]")
plt.xlabel("x [cm]"); plt.ylabel("y [cm]")
plt.title("Harta 2D e perturbimit magnetik")
plt.axis("equal"); plt.grid(True); plt.show()
```

### Interpretimi
- Harta 2D evidenton qartë zonën e perturbimit maksimal pranë magnetit me ΔB_max ≈ 117.3 µT.  
- Asimetria e lehtë e hartës reflekton orientimin jo perfekt të magnetit dhe sensorit.  
- Në skajet e hartës, fusha zbret gradualisht drejt vlerës ambientale.


| Eksperimenti | B_min [µT] | B_max [µT] | B_mean [µT] | DeltaB_max [µT] |
|--------------|------------|------------|-------------|-----------------|
| Sfondi       | 43.5       | 45.8       | 44.7        | -               |
| Magneti 1D   | 44.8       | 135.0      | 75.3        | 90.3            |
| Harta 2D     | 44.9       | 162.0      | 88.4        | 117.3           |
