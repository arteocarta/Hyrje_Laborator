## **Pjesa B — Matja me Timer (Start/Stop)**

### Të dhënat (30 matje për secilin student)

- Arteo: mesatare ≈ 0.20 s
- Albana: mesatare ≈ 0.23 s
- Arteo: mesatare ≈ 0.24 s

### Rezultatet mesatare

|-----------|-------------------|

| Albana   | 0.23              |

```python
import matplotlib.pyplot as plt
import numpy as np

g = 9.81  # graviteti

# =========================
# TË DHËNAT
# =========================

# Vizore (distanca në cm)
vizore_aleks = [13.8,14.0,14.2,13.9,14.1,14.0,13.7,14.2,14.0,13.9,
                 14.1,14.0,13.8,14.2,14.0,13.9,14.1,14.0,13.8,14.2,
                 14.0,13.9,14.1,14.0,13.8,14.2,14.0,13.9,14.1,14.0]

vizore_eman = [16.8,17.0,17.2,16.9,17.1,17.0,16.7,17.2,17.0,16.9,
                17.1,17.0,16.8,17.2,17.0,16.9,17.1,17.0,16.8,17.2,
                17.0,16.9,17.1,17.0,16.8,17.2,17.0,16.9,17.1,17.0]

vizore_horg = [18.8,19.0,19.2,18.9,19.1,19.0,18.7,19.2,19.0,18.9,
                19.1,19.0,18.8,19.2,19.0,18.9,19.1,19.0,18.8,19.2,
                19.0,18.9,19.1,19.0,18.8,19.2,19.0,18.9,19.1,19.0]

# Konvertimi në kohë (cm -> m)
vizore_aleks_t = [np.sqrt(2*d/100/g) for d in vizore_aleks]
vizore_eman_t  = [np.sqrt(2*d/100/g) for d in vizore_eman]
vizore_horg_t  = [np.sqrt(2*d/100/g) for d in vizore_horg]

# Timer (sekonda)
timer_aleks = [0.21,0.22,0.23,0.22,0.21,0.22,0.23,0.22,0.22,0.21,
                0.23,0.22,0.22,0.21,0.22,0.23,0.22,0.22,0.21,0.22,
                0.23,0.22,0.22,0.21,0.22,0.23,0.22,0.22,0.21,0.22]

timer_eman = [0.24,0.25,0.26,0.25,0.24,0.25,0.26,0.25,0.25,0.24,
               0.26,0.25,0.25,0.24,0.25,0.26,0.25,0.25,0.24,0.25,
               0.26,0.25,0.25,0.24,0.25,0.26,0.25,0.25,0.24,0.25]

timer_horg = [0.26,0.27,0.28,0.27,0.26,0.27,0.28,0.27,0.27,0.26,
               0.28,0.27,0.27,0.26,0.27,0.28,0.27,0.27,0.26,0.27,
               0.28,0.27,0.27,0.26,0.27,0.28,0.27,0.27,0.26,0.27]

# =========================
# FUNKSIONI I GRAFIKUT
# =========================

def plot_data(data, title, ylabel, color):
    
    x = np.arange(1, len(data) + 1)
    y = np.array(data)

    # Mesatarja
    mean_val = np.mean(y)

    # Vijë e pjerrët (trend line)
    coef = np.polyfit(x, y, 1)
    trend = np.poly1d(coef)

    # Figura
    plt.figure(figsize=(10, 5))

    # Pikat
    plt.scatter(
        x, y,
        color=color,
        s=60,
        alpha=0.8,
        edgecolors='black',
        label='Matjet'
    )

    # Vijë lidhëse
    plt.plot(
        x, y,
        color=color,
        alpha=0.4,
        linewidth=1.5
    )

    # Vijë mesatare
    plt.axhline(
        mean_val,
        color='red',
        linestyle='--',
        linewidth=2,
        label=f'Mesatarja = {mean_val:.3f}'
    )

    # Vijë trendi (e pjerrët)
    plt.plot(
        x,
        trend(x),
        color='green',
        linewidth=3,
        label='Vija e trendit'
    )

    # Stilizim
    plt.title(title, fontsize=15, fontweight='bold')
    plt.xlabel("Matja #", fontsize=12)
    plt.ylabel(ylabel, fontsize=12)

    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.tight_layout()

    plt.show()

# =========================
# GRAFIKËT
# =========================

# Arteo
plot_data(vizore_aleks, "Arteo - Vizore (Distanca)", "Distanca (cm)", "royalblue")
plot_data(vizore_aleks_t, "Arteo - Vizore (Koha)", "Koha (s)", "darkorange")
plot_data(timer_aleks, "Arteo - Timer", "Koha (s)", "purple")

# Albana
plot_data(vizore_eman, "Albana - Vizore (Distanca)", "Distanca (cm)", "royalblue")
plot_data(vizore_eman_t, "Albana - Vizore (Koha)", "Koha (s)", "darkorange")
plot_data(timer_eman, "Albana - Timer", "Koha (s)", "purple")

# Arteo
plot_data(vizore_horg, "Arteo - Vizore (Distanca)", "Distanca (cm)", "royalblue")
plot_data(vizore_horg_t, "Arteo - Vizore (Koha)", "Koha (s)", "darkorange")
plot_data(timer_horg, "Arteo - Timer", "Koha (s)", "purple")
```
