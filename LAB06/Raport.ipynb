## **Bazat Teorike**

### 1. Ç'është një matje?

Të matësh do të thotë të krahasosh madhësinë fizike me një njësi standarde. Asnjë matje nuk është absolutisht e saktë — rezultati shkruhet gjithmonë bashkë me pasigurinë e tij:

$$ x = x_{\text{mat}} \pm \Delta x $$

### 2. Klasifikimi i gabimeve

-   **Gabimet sistematike** → zhvendosin të gjitha matjet në të njëjtin drejtim (p.sh. instrument i pakalibruar)
-   **Gabimet rastësore** → luhatje statistikore rreth vlerës së vërtetë; zvogëlohen me përsëritjen e matjeve
-   **Gabimet trashanike** → pakujdesi e operatorit; këto nuk analizohen, por eliminohen

### 3. Rezolucioni dhe precizioni

-   **Rezolucioni** është ndarja më e vogël e shkallës së instrumentit
-   **Precizioni** tregon sa afër njëra-tjetrës grupohen matjet e përsëritura

Si rregull pune, pasiguria instrumentale vlerësohet sa **gjysma e ndarjes më të vogël** të shkallës.

## **Pjesa A — Matja e Gjatësisë**

### **Procedura**

U mat diametri i një cilindri alumini me tri instrumente me rezolucion të ndryshëm: vizore, kalibër vernier dhe mikrometër. Me secilin instrument u kryen nga 5 matje të pavarura, për të krahasuar precizionin e tyre.

### **Të dhënat**

| Instrument | Matja 1 | Matja 2 | Matja 3 | Matja 4 | Matja 5 | Mesatarja | Pasiguria |
|------------|---------|---------|---------|---------|---------|-----------|-----------|
| Vizore     | 25.2 mm | 25.1 mm | 25.3 mm | 25.2 mm | 25.1 mm | 25.18 mm  | ±0.5 mm   |
| Kalibër    | 25.14 mm| 25.16 mm| 25.15 mm| 25.14 mm| 25.16 mm| 25.15 mm  | ±0.02 mm  |
| Mikrometër | 25.147 mm|25.148 mm|25.146 mm|25.148 mm|25.147 mm| 25.147 mm | ±0.005 mm |

### **Analiza**

-   Mesatarja me vizore: $$ \bar{x} = 25.18 \, \text{mm} $$
-   Mesatarja me kalibër: $$ \bar{x} = 25.15 \, \text{mm} $$
-   Mesatarja me mikrometër: $$ \bar{x} = 25.147 \, \text{mm} $$

**Përfundim**: Sa më i imët rezolucioni i instrumentit, aq më e ngushtë del shpërndarja e matjeve. Mikrometri jep vlerën më të besueshme, ndërsa vizorja mbart pasigurinë më të madhe; megjithatë të tria mesataret përputhen brenda pasigurive përkatëse.

## **Pjesa B — Matja e Masës**

### **Procedura**

Masa e cilindrit të aluminit u përcaktua me peshore dixhitale. Matja u përsërit 5 herë, duke e hequr dhe rivendosur trupin mbi pjatën e peshores në çdo përsëritje.

### **Të dhënat**

| Matja 1 | Matja 2 | Matja 3 | Matja 4 | Matja 5 | Mesatarja | Pasiguria |
|---------|---------|---------|---------|---------|-----------|-----------|
| 33.52 g | 33.54 g | 33.53 g | 33.52 g | 33.53 g | 33.528 g  | ±0.01 g   |

### **Analiza**

-   Mesatarja: $$ \bar{m} = 33.528 \, \text{g} $$
-   Pasiguria: ±0.01 g (rezolucioni i peshores)
-   Leximet rezultuan të qëndrueshme dhe pa drift; luhatja rastësore është e krahasueshme me rezolucionin e instrumentit.

## **Pjesa C — Matja e Kohës**

### **Procedura**

U mat koha e rënies së lirë të një topi tenisi nga lartësia 1.2 m. Eksperimenti u përsërit 10 herë dhe koha u regjistrua me kronometër manual (smartphone).

### **Të dhënat**

| Matja 1 | Matja 2 | Matja 3 | Matja 4 | Matja 5 | Matja 6 | Matja 7 | Matja 8 | Matja 9 | Matja 10 |
|---------|---------|---------|---------|---------|---------|---------|---------|---------|----------|
| 0.50 s  | 0.49 s  | 0.51 s  | 0.48 s  | 0.50 s  | 0.51 s  | 0.49 s  | 0.50 s  | 0.52 s  | 0.48 s   |

### **Analiza**

-   Mesatarja: $$ \bar{t} = 0.498 \, \text{s} $$
-   Devijimi standard ≈ 0.013 s
-   Burimi mbizotërues i pasigurisë është **koha e reagimit të operatorit** gjatë nisjes dhe ndalimit manual të kronometrit.

**Krahasim me teorinë**:  
Nga formula $$ t = \sqrt{\frac{2h}{g}} $$  
me $$ h = 1.2 \, \text{m}, g = 9.81 \, \text{m/s}^2 $$  
del $$ t_{\text{teorik}} = 0.495 \, \text{s} $$  

→ Ndryshimi prej 0.003 s është dukshëm më i vogël se devijimi standard, pra rezultati eksperimental përputhet me parashikimin teorik brenda pasigurisë.
